## ðŸ”¹ Data Augmentation

Fine-grained image classification tasks, such as distinguishing between visually similar cat breeds, often suffer from issues like limited data diversity, class imbalance, and overfitting. To combat these challenges, this phase begins with the implementation of a comprehensive data augmentation pipeline that enhances both the quantity and variability of the training data without collecting new images.

### Why Data Augmentation is Critical:
- âœ… **Improves Generalization**: Helps the model learn robust, invariant features that perform well on unseen data.
- âœ… **Reduces Overfitting**: Introduces variation during training to prevent the model from memorizing specific features.
- âœ… **Simulates Real-World Conditions**: Mimics common visual distortions such as zoom, lighting changes, blur, and partial occlusion.
- âœ… **Boosts Performance on Fine-Grained Classes**: Encourages the model to focus on subtle, meaningful details rather than superficial cues.

### Augmentation Techniques Used:
The following transformations were composed into a `train` pipeline:

- **RandomResizedCrop (scale = 0.75â€“1.0)**  
  Simulates zooming and different object scales within the frame.
  
- **RandomHorizontalFlip**  
  Provides symmetry-based diversity for breeds with no directional bias.

- **ColorJitter (brightness, contrast, saturation = 0.4)**  
  Emulates lighting variations and photo conditions.

- **RandomAffine (rotation, translation, shear)**  
  Introduces positional and geometric diversity to the same breed.

- **GaussianBlur (kernel size = 3)**  
  Simulates motion blur or camera defocus, improving robustness to image quality variations.

- **AutoAugment (ImageNet Policy)**  
  Applies a learned sequence of high-level augmentations optimized on ImageNet to generalize well across datasets.

- **RandomErasing (p=0.2)**  
  Simulates occlusion or object obstruction (e.g., parts of the catâ€™s ear or tail not visible).

The `val` (validation) transform retains only resizing and normalization to ensure a fair and consistent evaluation pipeline.

Overall, this augmentation strategy plays a crucial role in enabling better **model generalization**, especially when fine-tuning deep CNNs on datasets with **high intra-class similarity** and **low inter-class variation**, like the 23-cat breed classification task.

```python
from torchvision.transforms import AutoAugment, AutoAugmentPolicy

data_transforms = {
    'train': transforms.Compose([
        transforms.RandomResizedCrop(224, scale=(0.75, 1.0)),  # Zoom effect
        transforms.RandomHorizontalFlip(),
        transforms.ColorJitter(brightness=0.4, contrast=0.4, saturation=0.4),
        transforms.RandomAffine(degrees=15, translate=(0.1, 0.1), scale=(0.9, 1.1), shear=10),
        transforms.GaussianBlur(kernel_size=3),
        AutoAugment(policy=AutoAugmentPolicy.IMAGENET),
        transforms.ToTensor(),
        transforms.Normalize(mean=[0.485, 0.456, 0.406],
                             std=[0.229, 0.224, 0.225]),
        transforms.RandomErasing(p=0.2, scale=(0.02, 0.1), ratio=(0.3, 3.3), value='random')
    ]),
    
    'val': transforms.Compose([
        transforms.Resize((224, 224)),
        transforms.ToTensor(),
        transforms.Normalize(mean=[0.485, 0.456, 0.406],
                             std=[0.229, 0.224, 0.225])
    ])
}
```

```python
dataset = datasets.ImageFolder(root=base_path, transform=data_transforms['train'])


# Plot: Transformed Images from Dataset
def show_transformed_images(dataset, num_images=10):
    plt.figure(figsize=(20, 5))
    
    for i in range(num_images):
        img, label = dataset[i]
        img = img.permute(1, 2, 0)  # Convert from (C, H, W) to (H, W, C)
        
        plt.subplot(1, num_images, i+1)
        plt.imshow(img)
        plt.title(dataset.classes[label])
        plt.axis('off')
    
    plt.tight_layout()
    plt.show()

# Visualize Transformed Images
show_transformed_images(dataset, num_images=5)
```

## ðŸ”¹ Fine-Tuning

With the augmented training dataset ready, the next step involves fine-tuning pretrained CNN models to adapt them to the fine-grained cat breed classification task. Fine-tuning enables models to go beyond generic ImageNet features and specialize in breed-specific visual cues.

For this experiment, two pretrained models were selected:
- **EfficientNetB0** â€” known for its performance and parameter efficiency through compound scaling.
- **MobileNetV2** â€” optimized for speed and compactness, suitable for mobile and edge deployment.

Each model is fine-tuned using multiple strategies that differ in:
- **The layers that are unfrozen** and updated during training
- **The learning rate scheduler** used to optimize convergence

These strategies allow us to explore how different levels of model flexibility and training dynamics affect final classification performance.

---

### ðŸ”¸ Fine-Tuning Strategy: EfficientNetB0

To effectively explore the learning capacity of EfficientNetB0, we define six progressively flexible fine-tuning strategies:

- **S1 & S2**: Only the classifier layer is trained.
- **S3**: The last two convolutional blocks and the classifier are fine-tuned.
- **S4**: All layers except the stem (`features[0]`) are fine-tuned â€” a near full fine-tuning configuration.
- **S5**: All layers of the model are unfrozen and retrained from top to bottom.
- **S6**: The final three convolutional blocks along with the classifier are fine-tuned.

The `prepare_efficientnet` function encapsulates this strategy logic, allowing consistent control over which layers are trainable. It also replaces the final classifier layer to output the correct number of classes for the cat breed dataset.

```python
def prepare_efficientnet(strategy, num_classes):
    model = models.efficientnet_b0(pretrained=True)

    # Freeze everything initially
    for param in model.parameters():
        param.requires_grad = False

    if strategy == 'S1' or strategy == 'S2':
        # Train classifier only
        for param in model.classifier.parameters():
            param.requires_grad = True

    elif strategy == 'S3':
        # Unfreeze last 2 blocks (features[6:], classifier)
        for name, param in model.features.named_parameters():
            if int(name.split('.')[0]) >= 6:
                param.requires_grad = True
        for param in model.classifier.parameters():
            param.requires_grad = True

    elif strategy == 'S4':
        # Unfreeze all except stem (features[0])
        for name, param in model.features.named_parameters():
            if not name.startswith('0.'):
                param.requires_grad = True
        for param in model.classifier.parameters():
            param.requires_grad = True

    elif strategy == 'S5':
        # Unfreeze everything
        for param in model.parameters():
            param.requires_grad = True

    elif strategy == 'S6':
        # Unfreeze last 3 blocks + classifier
        for name, param in model.features.named_parameters():
            if int(name.split('.')[0]) >= 5:
                param.requires_grad = True
        for param in model.classifier.parameters():
            param.requires_grad = True

    # Replace final classifier layer
    model.classifier[1] = nn.Linear(model.classifier[1].in_features, num_classes)

    return model
```

The following function defines the core training loop used across all fine-tuning strategies. It supports optional learning rate schedulers and tracks validation accuracy across epochs to identify the best-performing epoch.

- The model is trained using cross-entropy loss and optimized with `Adam`.
- If a scheduler is provided (e.g., StepLR, ReduceLROnPlateau), it adjusts learning rates dynamically.
- Validation accuracy is logged after each epoch for performance comparison across strategies.

```python
def train_model(model, train_loader, val_loader, criterion, optimizer, scheduler=None, strategy_name='', epochs=10):
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    model.to(device)

    val_acc = []
    best_acc = 0
    best_epoch = 0

    for epoch in range(epochs):
        model.train()
        for inputs, labels in train_loader:
            inputs, labels = inputs.to(device), labels.to(device)
            optimizer.zero_grad()
            outputs = model(inputs)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()

        # Validation
        model.eval()
        correct, total = 0, 0
        with torch.no_grad():
            for inputs, labels in val_loader:
                inputs, labels = inputs.to(device), labels.to(device)
                outputs = model(inputs)
                _, preds = torch.max(outputs, 1)
                total += labels.size(0)
                correct += (preds == labels).sum().item()

        acc = 100 * correct / total
        val_acc.append(acc)

        if acc > best_acc:
            best_acc = acc
            best_epoch = epoch + 1

        if scheduler:
            if isinstance(scheduler, torch.optim.lr_scheduler.ReduceLROnPlateau):
                scheduler.step(acc)
            else:
                scheduler.step()

        print(f"[{strategy_name}] Epoch {epoch+1}: Val Accuracy = {acc:.2f}%")

    print(f"[{strategy_name}] Best Val Accuracy = {best_acc:.2f}% at Epoch {best_epoch}")
    return model, val_acc
```

The loop below executes the fine-tuning process for all six EfficientNetB0 strategies (`S1` through `S6`). Each strategy is paired with a suitable learning rate scheduler, such as CosineAnnealing, StepLR, or OneCycleLR, to optimize training dynamics.

Validation accuracy is recorded for each epoch, and trained models are stored for later evaluation and comparison.

```python
from torch.optim.lr_scheduler import CosineAnnealingLR, StepLR, ReduceLROnPlateau, OneCycleLR, CosineAnnealingWarmRestarts

strategy_acc = {}
trained_models = {}

criterion = nn.CrossEntropyLoss()

for strategy in ['S1', 'S2', 'S3', 'S4', 'S5', 'S6']:
    model = prepare_efficientnet(strategy, num_classes=len(dataset.classes))
    optimizer = optim.Adam(filter(lambda p: p.requires_grad, model.parameters()), lr=0.001)

    # Define scheduler per strategy
    if strategy == 'S1':
        scheduler = None
    elif strategy == 'S2':
        scheduler = CosineAnnealingLR(optimizer, T_max=10)
    elif strategy == 'S3':
        scheduler = StepLR(optimizer, step_size=5, gamma=0.5)
    elif strategy == 'S4':
        scheduler = ReduceLROnPlateau(optimizer, patience=3, factor=0.3)
    elif strategy == 'S5':
        scheduler = OneCycleLR(optimizer, max_lr=0.001, steps_per_epoch=len(train_loader), epochs=10)
    elif strategy == 'S6':
        scheduler = CosineAnnealingWarmRestarts(optimizer, T_0=5)

    model, acc = train_model(
        model, train_loader, val_loader,
        criterion, optimizer, scheduler,
        strategy_name=strategy, epochs=10
    )

    strategy_acc[strategy] = acc
    trained_models[strategy] = model
```

#### ðŸ“Š Observations from EfficientNetB0 Fine-Tuning

- Strategies **S3**, **S4**, and **S6** demonstrated strong learning progression, achieving over **58% validation accuracy**, with **S6** peaking at **62.79%** â€” indicating that partial unfreezing of deeper layers paired with an advanced scheduler (CosineAnnealingWarmRestarts) can significantly improve performance.
- In contrast, **S1**, **S2**, and **S5** showed limited improvements, suggesting that shallow fine-tuning or aggressive full unfreezing without sufficient stabilization may hinder performance on fine-grained datasets.

While this reduces computational cost, it severely limits the modelâ€™s ability to adapt to the fine-grained nuances required to distinguish between visually similar cat breeds.

- **S5 (full fine-tuning)** surprisingly fails to outperform partial strategies. Although it has access to the entire model for updating, it likely suffers from **overfitting, unstable gradients, or suboptimal convergence**, especially with limited data and aggressive learning dynamics.

### ðŸ”¸ Fine-Tuning Strategy: MobileNetV2

MobileNetV2 was selected as the second backbone architecture due to its lightweight structure and high efficiency, particularly suitable for real-time or mobile deployments. Unlike EfficientNetB0, which uses compound scaling, MobileNetV2 adopts **depthwise separable convolutions** and **linear bottlenecks** to minimize computational overhead while preserving feature representation.

To evaluate the adaptability of MobileNetV2 to fine-grained classification tasks, we designed four custom fine-tuning strategies (M1â€“M4), each targeting a different subset of layers within the network. These strategies were tailored to investigate how different receptive field depths contribute to model performance.

The `prepare_mobilenet` function handles model configuration and selectively unfreezes layers based on the strategy:

- **M1**: Unfreezes only the final convolutional block (`features[17:]`) and the classifier. This is a conservative strategy intended to adapt high-level semantic features.

- **M2**: Fine-tunes from **block 13 onward**, including more of the deeper feature hierarchy. This aims to refine more abstract spatial features without altering the entire network.

- **M3**: Focuses on **mid-range bottleneck layers (blocks 7â€“13)**. This is designed to target mid-level features that may capture breed-specific textures or patterns.

- **M4**: Fully unfreezes the entire model, allowing all parameters â€” from early edge detectors to final semantics â€” to be retrained. This is the most aggressive strategy and is expected to either perform very well or suffer from overfitting depending on the optimizer and regularization.

All strategies conclude by replacing the final classification layer to match the 23 breed classes in the dataset.

```python
def prepare_mobilenet(strategy, num_classes):
    model = models.mobilenet_v2(pretrained=True)

    # Freeze all first
    for param in model.parameters():
        param.requires_grad = False

    if strategy == 'M1':
        # Unfreeze last conv + classifier
        for name, param in model.features.named_parameters():
            if int(name.split('.')[0]) >= 17:
                param.requires_grad = True
        for param in model.classifier.parameters():
            param.requires_grad = True

    elif strategy == 'M2':
        # Unfreeze from feature block 13 onward
        for name, param in model.features.named_parameters():
            if int(name.split('.')[0]) >= 13:
                param.requires_grad = True
        for param in model.classifier.parameters():
            param.requires_grad = True

    elif strategy == 'M3':
        # Unfreeze middle bottlenecks: 7â€“13
        for name, param in model.features.named_parameters():
            if 7 <= int(name.split('.')[0]) <= 13:
                param.requires_grad = True
        for param in model.classifier.parameters():
            param.requires_grad = True

    elif strategy == 'M4':
        # Unfreeze all layers
        for param in model.parameters():
            param.requires_grad = True

    # Replace final classifier
    model.classifier[1] = nn.Linear(model.classifier[1].in_features, num_classes)

    return model
```

```python
from torch.optim.lr_scheduler import ExponentialLR, CyclicLR

mobilenet_acc = {}
mobilenet_models = {}

for strategy in ['M1', 'M2', 'M3', 'M4']:
    model = prepare_mobilenet(strategy, num_classes=len(dataset.classes))
    optimizer = optim.Adam(filter(lambda p: p.requires_grad, model.parameters()), lr=0.001)

    # Assign scheduler
    if strategy == 'M1':
        scheduler = CosineAnnealingWarmRestarts(optimizer, T_0=5)
    elif strategy == 'M2':
        scheduler = ExponentialLR(optimizer, gamma=0.9)
    elif strategy == 'M3':
        scheduler = CyclicLR(optimizer, base_lr=1e-5, max_lr=1e-3, step_size_up=5, mode='triangular')
    elif strategy == 'M4':
        scheduler = OneCycleLR(optimizer, max_lr=0.001, steps_per_epoch=len(train_loader), epochs=10)

    model, acc = train_model(
        model, train_loader, val_loader,
        criterion, optimizer, scheduler,
        strategy_name=strategy, epochs=10
    )

    mobilenet_acc[strategy] = acc
    mobilenet_models[strategy] = model
```

#### ðŸ“Š Observations from MobileNetV2 Fine-Tuning

- **Strategy M1** performed the best among all MobileNetV2 configurations, reaching a peak validation accuracy of **53.49%**, indicating that fine-tuning the last convolutional block and classifier is sufficient to adapt high-level features to the target domain.

- **M2** also showed competitive results, achieving **48.84%** at epoch 7, suggesting that unfreezing a slightly larger portion of deeper feature layers can enhance performance, though it showed less stability across epochs.

- **M3** and **M4** underperformed significantly. M3, which focused only on mid-level bottlenecks, struggled to converge, likely due to the lack of support from high-level feature adaptation. M4, which fully unfroze the entire network, suffered from noisy gradients and unstable learning, highlighting that **aggressive full fine-tuning on limited data may hinder performance** without careful regularization.

These results reinforce that **targeted fine-tuning of semantically rich layers** provides the best balance between model flexibility and stability in MobileNetV2.

The performance gap between M1/M2 and M3/M4 indicates that efficient transfer learning in lightweight networks benefits more from **targeted flexibility** than full retraining.

## ðŸ”¹ Model Evaluation & Comparison

The final stage of this phase involves evaluating the performance of all fine-tuned models on the **unseen test dataset**. This step is crucial for measuring how well each model generalizes beyond the training and validation sets, particularly in a fine-grained classification scenario where breed-level distinctions are often subtle and visually similar.

### Evaluation Objective:

While validation accuracy provided insight into learning progress and convergence, **test accuracy offers an unbiased estimate** of each model's real-world performance. It reflects how well the learned features and classification layers can generalize to new, unobserved data.

### Evaluation Process:

- A common evaluation function (`evaluate_test_accuracy`) was used across all models to ensure consistency.
- Each model was evaluated on the same **test_loader**, containing data that was never seen during training or validation.
- The final **test accuracy** was calculated as the proportion of correctly classified images out of the total number of test samples.
- A total of **10 models** were evaluated â€” **6 EfficientNetB0 strategies (S1â€“S6)** and **4 MobileNetV2 strategies (M1â€“M4)**.
- The results were recorded and will be visualized to facilitate comparison.

This evaluation provides a direct basis for identifying which fine-tuning strategy is most effective for **fine-grained cat breed classification**, and whether deeper or more conservative transfer learning techniques yield better generalization.

```python
def evaluate_test_accuracy(model, test_loader):
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    model.to(device)
    model.eval()

    correct, total = 0, 0
    with torch.no_grad():
        for inputs, labels in test_loader:
            inputs, labels = inputs.to(device), labels.to(device)
            outputs = model(inputs)
            _, preds = torch.max(outputs, 1)
            correct += (preds == labels).sum().item()
            total += labels.size(0)

    return 100 * correct / total
```

```python
all_test_accuracies = {}
all_models = {**trained_models, **mobilenet_models}

for name, model in all_models.items():
    acc = evaluate_test_accuracy(model, test_loader)
    all_test_accuracies[name] = acc
    print(f"{name} Test Accuracy: {acc:.2f}%")
```

### 2. Misclassified Image Analysis and Common Patterns

To better understand the limitations of the models beyond numeric accuracy scores, we conducted a visual inspection of misclassified test samples. By examining prediction errors, we can identify recurring confusion patterns and potential failure points.

The visualization function below selects and displays a subset of incorrectly classified images for a given model. It overlays each image with its true and predicted labels, providing immediate insight into where the model struggles and why.

This analysis helps in:

- Highlighting **visually similar breed confusions**, such as Persian vs. Exotic or Sphynx vs. Siamese, which often share fur textures, pose, or facial structure.
- Revealing misclassifications caused by **occlusion, blur, low lighting**, or atypical poses, where key visual cues are either missing or distorted.
- Informing potential improvements such as **attention mechanisms, multi-view learning**, or **additional augmentation** to make the model more robust to real-world conditions.

By examining these misclassified samples, we go beyond accuracy scores to explore the **semantic behavior** of the models and pinpoint the **most critical areas for improvement** in future iterations.

```python
def visualize_misclassified(model, test_loader, class_names, num_images=10):
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    model.to(device)
    model.eval()

    misclassified = []

    with torch.no_grad():
        for inputs, labels in test_loader:
            inputs, labels = inputs.to(device), labels.to(device)
            outputs = model(inputs)
            _, preds = torch.max(outputs, 1)

            for i in range(len(labels)):
                if preds[i] != labels[i]:
                    misclassified.append((inputs[i].cpu(), labels[i].cpu(), preds[i].cpu()))
                if len(misclassified) >= num_images:
                    break
            if len(misclassified) >= num_images:
                break

    # Plot
    plt.figure(figsize=(20, 5))
    for idx, (img, true, pred) in enumerate(misclassified):
        img = img.permute(1, 2, 0).numpy()
        img = img * np.array([0.229, 0.224, 0.225]) + np.array([0.485, 0.456, 0.406])
        img = np.clip(img, 0, 1)

        plt.subplot(1, num_images, idx + 1)
        plt.imshow(img)
        plt.title(f"True: {class_names[true]}\nPred: {class_names[pred]}", fontsize=9)
        plt.axis('off')
    plt.tight_layout()
    plt.show()
```

```python
class_names = dataset.classes

print("Misclassified Images â€” EfficientNetB0 (S4)")
visualize_misclassified(all_models['S4'], test_loader, class_names)

print("Misclassified Images â€” MobileNetV2 (M2)")
visualize_misclassified(all_models['M2'], test_loader, class_names)
```

```python
from collections import Counter, defaultdict

def get_misclassification_patterns(model, test_loader, class_names):
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    model.to(device)
    model.eval()

    misclassified_pairs = []

    with torch.no_grad():
        for inputs, labels in test_loader:
            inputs, labels = inputs.to(device), labels.to(device)
            outputs = model(inputs)
            _, preds = torch.max(outputs, 1)

            for true, pred in zip(labels, preds):
                if true != pred:
                    true_class = class_names[true]
                    pred_class = class_names[pred]
                    misclassified_pairs.append((true_class, pred_class))

    counter = Counter(misclassified_pairs)
    return counter
```

```python
patterns_s4 = get_misclassification_patterns(all_models['S4'], test_loader, class_names)
patterns_m2 = get_misclassification_patterns(all_models['M2'], test_loader, class_names)

print("Top 10 Confusions â€“ EfficientNetB0 (S4)")
for pair, count in patterns_s4.most_common(10):
    print(f"{pair[0]} â†’ {pair[1]}: {count} times")

print("\nTop 10 Confusions â€“ MobileNetV2 (M2)")
for pair, count in patterns_m2.most_common(10):
    print(f"{pair[0]} â†’ {pair[1]}: {count} times")
```

```python
import matplotlib.pyplot as plt
import numpy as np

def plot_confusion_pair_groupedbar(counter1, counter2, title1, title2, top_n=10):
    # Get top N confusions
    top1 = counter1.most_common(top_n)
    top2 = counter2.most_common(top_n)

    labels1 = [f"{true}â†’{pred}" for (true, pred), _ in top1]
    counts1 = [count for _, count in top1]

    labels2 = [f"{true}â†’{pred}" for (true, pred), _ in top2]
    counts2 = [count for _, count in top2]

    # Align by label position
    x = np.arange(top_n)
    width = 0.35

    fig, ax = plt.subplots(figsize=(14, 6))
    bars1 = ax.bar(x - width/2, counts1, width, label=title1, color='steelblue')
    bars2 = ax.bar(x + width/2, counts2, width, label=title2, color='salmon')

    # Add value annotations
    for bar in bars1 + bars2:
        height = bar.get_height()
        ax.annotate(f'{height}',
                    xy=(bar.get_x() + bar.get_width() / 2, height),
                    xytext=(0, 3), textcoords="offset points",
                    ha='center', va='bottom', fontsize=9)

    ax.set_xlabel('Confusion Pairs (True â†’ Predicted)', fontsize=12)
    ax.set_ylabel('Count', fontsize=12)
    ax.set_title('Top 10 Misclassifications', fontsize=14)
    ax.set_xticks(x)
    ax.set_xticklabels(labels1, rotation=45, ha='right')
    ax.legend()
    ax.grid(axis='y', linestyle='--', alpha=0.6)
    plt.tight_layout()
    plt.show()
```

```python
plot_confusion_pair_groupedbar(
    patterns_s4,
    patterns_m2,
    "EfficientNetB0 (S4)",
    "MobileNetV2 (M2)"
)
```
