```python
# Imports
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

import torch
import torchvision
from torchvision import datasets, transforms, models
from torch.utils.data import DataLoader, random_split
import torch.nn as nn
import torch.optim as optim
from sklearn.metrics import confusion_matrix, classification_report
```

```python
# Dataset Path for Subhana
base_path = 'Images and labels/Split_By_Members/Subhana'

# Labels CSV for Subhana
labels_df = pd.read_csv('Images and labels/subhana_labels.csv')

# Check basic info
print("Total Images:", len(labels_df))
print("Breeds Available:", labels_df['Breed Name'].nunique())

labels_df.head()
```

```python
plt.figure(figsize=(15, 6))
sns.countplot(data=labels_df, x='Breed Name', order=labels_df['Breed Name'].value_counts().index)
plt.xticks(rotation=90)
plt.title('Images per Breed in Subhana\'s Dataset')
plt.show()
```

```python
from PIL import Image
import random

plt.figure(figsize=(12, 12))

sample_images = labels_df.sample(9)

for idx, row in enumerate(sample_images.iterrows()):
    relative_path = row[1]['Path']
    full_path = os.path.join('Images and labels', relative_path)

    if os.path.exists(full_path):
        img = Image.open(full_path)
        plt.subplot(3, 3, idx+1)
        plt.imshow(img)
        plt.title(row[1]['Breed Name'])
        plt.axis('off')
    else:
        print(f"Image Missing: {full_path}")

plt.tight_layout()
plt.show()
```

```python
from torchvision import transforms

# Define Train & Validation/Test Transforms
data_transforms = {
    'train': transforms.Compose([
        transforms.Resize((224, 224)),
        transforms.RandomHorizontalFlip(),
        transforms.RandomRotation(10),
        transforms.ToTensor(),
        transforms.Normalize([0.485, 0.456, 0.406],
                             [0.229, 0.224, 0.225])
    ]),
    'val': transforms.Compose([
        transforms.Resize((224, 224)),
        transforms.ToTensor(),
        transforms.Normalize([0.485, 0.456, 0.406],
                             [0.229, 0.224, 0.225])
    ]),
}
```

```python
from torchvision import datasets
from torch.utils.data import DataLoader, random_split

# Load Dataset
dataset = datasets.ImageFolder(root=base_path, transform=data_transforms['train'])

# Check Class to Index Mapping
print("Breed to Label Mapping:\n")
print(dataset.class_to_idx)

print(f"\nTotal Images in Dataset: {len(dataset)}")
```

```python
# Calculate Split Sizes
train_size = int(0.6 * len(dataset))
val_size = int(0.1 * len(dataset))
test_size = len(dataset) - train_size - val_size

# Perform Random Split
train_data, val_data, test_data = random_split(dataset, [train_size, val_size, test_size])

print(f"Train Images: {len(train_data)}")
print(f"Validation Images: {len(val_data)}")
print(f"Test Images: {len(test_data)}")
```

```python
from torch.utils.data import DataLoader

# Batch size for loading
batch_size = 32

# DataLoaders
train_loader = DataLoader(train_data, batch_size=batch_size, shuffle=True)
val_loader = DataLoader(val_data, batch_size=batch_size, shuffle=False)
test_loader = DataLoader(test_data, batch_size=batch_size, shuffle=False)

print("DataLoaders Created Successfully!")
```

```python
# Load Pretrained VGG16
vgg_model = models.vgg16(pretrained=True)

# Freeze Feature Extractor Layers
for param in vgg_model.features.parameters():
    param.requires_grad = False

# Modify Classifier for 23 Breeds
vgg_model.classifier[6] = nn.Linear(4096, 23)

# Move Model to Device (GPU if available)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
vgg_model = vgg_model.to(device)

print(vgg_model)
```

```python
# Loss Function
criterion = nn.CrossEntropyLoss()

# Optimizer - Only training the classifier parameters
optimizer = optim.Adam(vgg_model.classifier.parameters(), lr=0.001)

print("Loss Function & Optimizer set successfully!")
```

```python
def train_model(model, train_loader, val_loader, criterion, optimizer, epochs=10):
    for epoch in range(epochs):
        model.train()
        running_loss = 0.0
        
        for inputs, labels in train_loader:
            inputs, labels = inputs.to(device), labels.to(device)

            optimizer.zero_grad()
            outputs = model(inputs)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()
            
            running_loss += loss.item()

        print(f"Epoch [{epoch+1}/{epochs}] | Train Loss: {running_loss/len(train_loader):.4f}")

    print("Training Completed")
```

```python
def evaluate_model(model, test_loader):
    model.eval()
    correct = 0
    total = 0
    all_preds = []
    all_labels = []

    with torch.no_grad():
        for inputs, labels in test_loader:
            inputs, labels = inputs.to(device), labels.to(device)
            outputs = model(inputs)
            _, preds = torch.max(outputs, 1)

            all_preds.extend(preds.cpu().numpy())
            all_labels.extend(labels.cpu().numpy())

            correct += (preds == labels).sum().item()
            total += labels.size(0)

    print(f"Test Accuracy: {100 * correct / total:.2f}%")
    return all_labels, all_preds
```

```python
# Start Training
train_model(vgg_model, train_loader, val_loader, criterion, optimizer, epochs=10)
```

```python
def evaluate_model(model, test_loader):
    model.eval()
    correct = 0
    total = 0
    all_preds = []
    all_labels = []

    with torch.no_grad():
        for inputs, labels in test_loader:
            inputs, labels = inputs.to(device), labels.to(device)
            outputs = model(inputs)
            _, preds = torch.max(outputs, 1)

            all_preds.extend(preds.cpu().numpy())
            all_labels.extend(labels.cpu().numpy())

            correct += (preds == labels).sum().item()
            total += labels.size(0)

    accuracy = 100 * correct / total
    print(f"Test Accuracy: {accuracy:.2f}%")
    return all_labels, all_preds
```

```python
# Evaluate VGG16
true_labels, predictions = evaluate_model(vgg_model, test_loader)
```

```python
# Generate Classification Report
print(classification_report(true_labels, predictions, target_names=dataset.classes))
```

```python
cm = confusion_matrix(true_labels, predictions)

plt.figure(figsize=(15, 10))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=dataset.classes,
            yticklabels=dataset.classes)

plt.title("Confusion Matrix - VGG16")
plt.xlabel("Predicted")
plt.ylabel("True")
plt.show()
```

```python
# Load Pretrained ResNet50
resnet_model = models.resnet50(pretrained=True)

# Freeze Feature Extractor Layers
for param in resnet_model.parameters():
    param.requires_grad = False

# Modify Final Fully Connected Layer for 23 Classes
num_features = resnet_model.fc.in_features
resnet_model.fc = nn.Linear(num_features, 23)

# Move Model to Device
resnet_model = resnet_model.to(device)

print(resnet_model)
```

```python
# Loss Function
criterion_resnet = nn.CrossEntropyLoss()

# Optimizer - Only train the final FC layer
optimizer_resnet = optim.Adam(resnet_model.fc.parameters(), lr=0.001)

print("Loss Function & Optimizer for ResNet50 are ready!")
```

```python
# Train Function already defined as train_model()

# Train ResNet50
train_model(resnet_model, train_loader, val_loader, criterion_resnet, optimizer_resnet, epochs=10)
```

```python
# Evaluate ResNet50
true_labels_resnet, predictions_resnet = evaluate_model(resnet_model, test_loader)
```

```python
# Classification Report for ResNet50
print(classification_report(true_labels_resnet, predictions_resnet, target_names=dataset.classes))
```

```python
cm_resnet = confusion_matrix(true_labels_resnet, predictions_resnet)

plt.figure(figsize=(15, 10))
sns.heatmap(cm_resnet, annot=True, fmt='d', cmap='Greens',
            xticklabels=dataset.classes,
            yticklabels=dataset.classes)

plt.title("Confusion Matrix - ResNet50")
plt.xlabel("Predicted")
plt.ylabel("True")
plt.show()
```

```python
# Accuracy Calculation Function
def calculate_accuracy(true, preds):
    correct = sum(np.array(true) == np.array(preds))
    total = len(true)
    return 100 * correct / total

# Calculate Accuracy
vgg_acc = calculate_accuracy(true_labels, predictions)
resnet_acc = calculate_accuracy(true_labels_resnet, predictions_resnet)

print(f"VGG16 Accuracy  : {vgg_acc:.2f}%")
print(f"ResNet50 Accuracy: {resnet_acc:.2f}%")
```

```python
print("Model Comparison Summary:\n")
print(f"VGG16 achieved an accuracy of {vgg_acc:.2f}% on Subhana's test dataset.")
print(f"ResNet50 achieved an accuracy of {resnet_acc:.2f}% on the same dataset.")

if vgg_acc > resnet_acc:
    print("\nObservation: VGG16 performed slightly better in this case.")
else:
    print("\nObservation: ResNet50 performed slightly better in this case.")

print("Further tuning, more data, or fine-tuning full model layers can improve performance.")
```
