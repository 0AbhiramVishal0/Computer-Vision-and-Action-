`````python
# Basic Libraries
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

# PyTorch Libraries
import torch
import torchvision
from torchvision import datasets, transforms, models
from torch.utils.data import DataLoader, random_split
import torch.nn as nn
import torch.optim as optim


from sklearn.metrics import confusion_matrix, classification_report
from PIL import Image
import random
`````

`````python
base_path = 'Images and labels/All_Breeds_Dataset'
labels_df = pd.read_csv('Images and labels/final_labels.csv')

labels_df['Path'] = labels_df['Path'].apply(lambda x: os.path.join("Images and labels", x.replace('\\', '/')))

print("Total Images:", len(labels_df))
print("Total Breeds:", labels_df['Breed Name'].nunique())
labels_df.head()
`````

`````python
from torch.utils.data import Dataset
from PIL import Image
import os

class CatBreedDataset(Dataset):
    def __init__(self, df, transform=None):
        self.df = df.copy()
        self.transform = transform
        self.classes = sorted(df['Breed Name'].unique())
        self.class_to_idx = {c: i for i, c in enumerate(self.classes)}
    
    def __len__(self):
        return len(self.df)
    
    def __getitem__(self, idx):
        row = self.df.iloc[idx]
        image_path = row['Path']  # Already correct relative path
        image = Image.open(image_path).convert('RGB')
        label = self.class_to_idx[row['Breed Name']]
        if self.transform:
            image = self.transform(image)
        return image, label
`````

The following data augmentation and preprocessing transformations are applied to the training and validation datasets to improve model generalization.

`````python
from torchvision import transforms
from torchvision.transforms import AutoAugment, AutoAugmentPolicy

data_transforms = {
    'train': transforms.Compose([
        transforms.RandomResizedCrop(224, scale=(0.75, 1.0)),
        transforms.RandomHorizontalFlip(),
        transforms.ColorJitter(0.4, 0.4, 0.4),
        transforms.RandomAffine(degrees=15, translate=(0.1, 0.1), scale=(0.9, 1.1), shear=10),
        transforms.GaussianBlur(kernel_size=3),
        AutoAugment(policy=AutoAugmentPolicy.IMAGENET),
        transforms.ToTensor(),
        transforms.Normalize(mean=[0.485, 0.456, 0.406],
                             std=[0.229, 0.224, 0.225]),
        transforms.RandomErasing(p=0.2, scale=(0.02, 0.1), ratio=(0.3, 3.3), value='random')
    ]),
    'val': transforms.Compose([
        transforms.Resize((224, 224)),
        transforms.ToTensor(),
        transforms.Normalize(mean=[0.485, 0.456, 0.406],
                             std=[0.229, 0.224, 0.225])
    ])
}
`````

`````python
from torch.utils.data import DataLoader, random_split

full_dataset = CatBreedDataset(labels_df, transform=data_transforms['train'])
train_size = int(0.6 * len(full_dataset))
val_size = int(0.1 * len(full_dataset))
test_size = len(full_dataset) - train_size - val_size
train_data, val_data, test_data = random_split(full_dataset, [train_size, val_size, test_size])

batch_size = 32
train_loader = DataLoader(train_data, batch_size=batch_size, shuffle=True)
val_loader = DataLoader(val_data, batch_size=batch_size, shuffle=True)
test_loader = DataLoader(test_data, batch_size=batch_size, shuffle=False)
`````

`````python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
criterion = nn.CrossEntropyLoss()
num_classes = len(full_dataset.classes)
`````

The following functions define the ResNet50 architecture with Strategy B2 and implement the corresponding training loop. The training procedure uses the specified optimizer, loss function, and learning rate scheduler, and evaluates model performance on the validation set after each epoch.


`````python
def prepare_resnet50(strategy='B'):
    model = models.resnet50(pretrained=True)

    if strategy == 'B':
        for param in model.parameters():
            param.requires_grad = False
        for param in model.layer4.parameters():
            param.requires_grad = True
        for param in model.fc.parameters():
            param.requires_grad = True

    model.fc = nn.Linear(model.fc.in_features, num_classes)
    return model

def train_resnet(model, train_loader, val_loader, criterion, optimizer, scheduler=None, epochs=10):
    model.to(device)
    val_acc = []
    best_acc = 0
    best_epoch = 0

    for epoch in range(epochs):
        model.train()
        for inputs, labels in train_loader:
            inputs, labels = inputs.to(device), labels.to(device)
            optimizer.zero_grad()
            outputs = model(inputs)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()

        # Validation
        model.eval()
        correct, total = 0, 0
        with torch.no_grad():
            for inputs, labels in val_loader:
                inputs, labels = inputs.to(device), labels.to(device)
                outputs = model(inputs)
                _, preds = torch.max(outputs, 1)
                total += labels.size(0)
                correct += (preds == labels).sum().item()

        acc = 100 * correct / total
        val_acc.append(acc)

        if acc > best_acc:
            best_acc = acc
            best_epoch = epoch + 1

        if scheduler:
            if isinstance(scheduler, torch.optim.lr_scheduler.ReduceLROnPlateau):
                scheduler.step(acc)
            else:
                scheduler.step()

        print(f"[ResNet50-B2] Epoch {epoch+1}: Val Accuracy = {acc:.2f}%")

    print(f"[ResNet50-B2] Best Val Accuracy = {best_acc:.2f}% at Epoch {best_epoch}")
    return model, val_acc
`````

The following functions define the VGG16 architecture with Strategy B1, where selected convolutional layers and the classifier are unfrozen for fine-tuning. The training loop applies the optimizer, loss function, and learning rate scheduler while tracking validation accuracy across epochs. This enables evaluation of the model’s ability to learn fine-grained cat breed features from the full dataset.


`````python
def prepare_vgg16(strategy='B'):
    model = models.vgg16(pretrained=True)

    if strategy == 'B':
        for param in model.parameters():
            param.requires_grad = False
        for param in model.features[24:].parameters():
            param.requires_grad = True
        for param in model.classifier.parameters():
            param.requires_grad = True

    model.classifier[6] = nn.Linear(model.classifier[6].in_features, num_classes)
    return model

def train_vgg(model, train_loader, val_loader, criterion, optimizer, scheduler=None, epochs=10):
    model.to(device)
    val_acc = []

    for epoch in range(epochs):
        model.train()
        for inputs, labels in train_loader:
            inputs, labels = inputs.to(device), labels.to(device)
            optimizer.zero_grad()
            outputs = model(inputs)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()

        # Validation
        model.eval()
        correct, total = 0, 0
        with torch.no_grad():
            for inputs, labels in val_loader:
                inputs, labels = inputs.to(device), labels.to(device)
                outputs = model(inputs)
                _, preds = torch.max(outputs, 1)
                total += labels.size(0)
                correct += (preds == labels).sum().item()

        acc = 100 * correct / total
        val_acc.append(acc)

        if scheduler:
            if isinstance(scheduler, torch.optim.lr_scheduler.ReduceLROnPlateau):
                scheduler.step(acc)
            else:
                scheduler.step()

        print(f"[VGG16-B1] Epoch {epoch+1}: Val Accuracy = {acc:.2f}%")

    best_acc = max(val_acc)
    best_epoch = np.argmax(val_acc) + 1
    print(f"[VGG16-B1] Best Val Accuracy = {best_acc:.2f}% at Epoch {best_epoch}")
    return model, val_acc
`````

The following functions define the EfficientNetB0 architecture with Strategy S6, where the last three feature blocks and the classifier are unfrozen for fine-tuning. The training loop monitors validation performance while applying optimization and learning rate scheduling. This approach aims to leverage EfficientNetB0’s efficient feature extraction while adapting it to the fine-grained characteristics of the full cat breed dataset.


`````python
def prepare_efficientnet(strategy='S6'):
    model = models.efficientnet_b0(pretrained=True)

    # Freeze all first
    for param in model.parameters():
        param.requires_grad = False

    # Unfreeze last 3 blocks + classifier
    for name, param in model.features.named_parameters():
        if int(name.split('.')[0]) >= 5:
            param.requires_grad = True
    for param in model.classifier.parameters():
        param.requires_grad = True

    model.classifier[1] = nn.Linear(model.classifier[1].in_features, num_classes)
    return model

# Use the same train_model function as EfficientNet phase2:
def train_model(model, train_loader, val_loader, criterion, optimizer, scheduler=None, strategy_name='', epochs=10):
    model.to(device)
    val_acc = []
    best_acc = 0
    best_epoch = 0

    for epoch in range(epochs):
        model.train()
        for inputs, labels in train_loader:
            inputs, labels = inputs.to(device), labels.to(device)
            optimizer.zero_grad()
            outputs = model(inputs)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()

        # Validation
        model.eval()
        correct, total = 0, 0
        with torch.no_grad():
            for inputs, labels in val_loader:
                inputs, labels = inputs.to(device), labels.to(device)
                outputs = model(inputs)
                _, preds = torch.max(outputs, 1)
                total += labels.size(0)
                correct += (preds == labels).sum().item()

        acc = 100 * correct / total
        val_acc.append(acc)

        if acc > best_acc:
            best_acc = acc
            best_epoch = epoch + 1

        if scheduler:
            if isinstance(scheduler, torch.optim.lr_scheduler.ReduceLROnPlateau):
                scheduler.step(acc)
            else:
                scheduler.step()

        print(f"[{strategy_name}] Epoch {epoch+1}: Val Accuracy = {acc:.2f}%")

    print(f"[{strategy_name}] Best Val Accuracy = {best_acc:.2f}% at Epoch {best_epoch}")
    return model, val_acc
`````

The following function defines the MobileNetV2 architecture with Strategy M1, where the last convolutional block and the classifier are unfrozen for fine-tuning. This strategy balances computational efficiency with the model’s ability to adapt to the fine-grained cat breed classification task, making it suitable for deployment on resource-constrained platforms.

`````python
def prepare_mobilenet(strategy='M1'):
    model = models.mobilenet_v2(pretrained=True)

    # Freeze all first
    for param in model.parameters():
        param.requires_grad = False

    # Unfreeze last conv + classifier
    for name, param in model.features.named_parameters():
        if int(name.split('.')[0]) >= 17:
            param.requires_grad = True
    for param in model.classifier.parameters():
        param.requires_grad = True

    model.classifier[1] = nn.Linear(model.classifier[1].in_features, num_classes)
    return model
`````

The selected models are trained using the full Phase 2 dataset with consistent training configurations. Each model is optimized with the Adam optimizer and a ReduceLROnPlateau learning rate scheduler. The training process tracks validation accuracy over ten epochs, enabling direct comparison of model performance for final selection.

`````python
# ResNet50-B2
resnet_model = prepare_resnet50('B')
opt_resnet = optim.Adam(filter(lambda p: p.requires_grad, resnet_model.parameters()), lr=0.0005)
sched_resnet = optim.lr_scheduler.ReduceLROnPlateau(opt_resnet, patience=3, factor=0.3)
resnet_model, resnet_acc = train_resnet(resnet_model, train_loader, val_loader, criterion, opt_resnet, sched_resnet)

# VGG16-B1
vgg_model = prepare_vgg16('B')
opt_vgg = optim.Adam(filter(lambda p: p.requires_grad, vgg_model.parameters()), lr=0.0005)
sched_vgg = optim.lr_scheduler.ReduceLROnPlateau(opt_vgg, patience=3, factor=0.3)
vgg_model, vgg_acc = train_vgg(vgg_model, train_loader, val_loader, criterion, opt_vgg, sched_vgg)

# EfficientNetB0-S6
effnet_model = prepare_efficientnet('S6')
opt_effnet = optim.Adam(filter(lambda p: p.requires_grad, effnet_model.parameters()), lr=0.001)
sched_effnet = optim.lr_scheduler.ReduceLROnPlateau(opt_effnet, patience=3, factor=0.3)
effnet_model, effnet_acc = train_model(effnet_model, train_loader, val_loader, criterion, opt_effnet, sched_effnet, strategy_name='EfficientNetB0-S6')

# MobileNetV2-M1
mobilenet_model = prepare_mobilenet('M1')
opt_mobilenet = optim.Adam(filter(lambda p: p.requires_grad, mobilenet_model.parameters()), lr=0.001)
sched_mobilenet = optim.lr_scheduler.ReduceLROnPlateau(opt_mobilenet, patience=3, factor=0.3)
mobilenet_model, mobilenet_acc = train_model(mobilenet_model, train_loader, val_loader, criterion, opt_mobilenet, sched_mobilenet, strategy_name='MobileNetV2-M1')
`````

`````python
# Plot validation accuracy comparison
plt.figure(figsize=(10,6))
plt.plot(resnet_acc, label='ResNet50-B2')
plt.plot(vgg_acc, label='VGG16-B1')
plt.plot(effnet_acc, label='EfficientNetB0-S6')
plt.plot(mobilenet_acc, label='MobileNetV2-M1')
plt.xlabel("Epoch")
plt.ylabel("Validation Accuracy (%)")
plt.title("Validation Accuracy Comparison")
plt.legend()
plt.grid()
plt.show()
`````

### Model Performance Analysis and Selection

The validation accuracy comparison plot provides the following observations:

- **EfficientNetB0-S6**
  - Initial accuracy: ~37%
  - Rapid improvement through the first 5 epochs, reaching ~60% by epoch 5
  - Final validation accuracy: **~70%**
  - Very stable learning curve with minimal oscillation in later epochs
  - Highest overall accuracy across all models

- **ResNet50-B2**
  - Initial accuracy: ~36%
  - Steady improvement, reaching ~53% by epoch 5
  - Final validation accuracy: **~65%**
  - Slight oscillation observed between epochs 7–10
  - Second-best overall accuracy, but higher computational cost

- **MobileNetV2-M1**
  - Initial accuracy: ~42%
  - Gradual improvement to **~63%** final accuracy
  - Smooth and stable learning curve
  - Lower accuracy than EfficientNetB0-S6 and ResNet50-B2

- **VGG16-B1**
  - Initial accuracy: ~13%
  - Slow progression throughout training
  - Final validation accuracy: **~55%**
  - Clearly underperforms compared to other models

**Model selection rationale:**

EfficientNetB0-S6 achieves the **highest final validation accuracy (~70%)**, while also providing a stable and consistent learning curve. In addition to accuracy, EfficientNetB0-B6 is an architecture specifically designed for efficient inference, making it suitable for deployment on resource-constrained systems such as the robot platform. Although ResNet50-B2 achieves competitive accuracy (~65%), its computational demands are significantly higher. MobileNetV2-M1, while efficient, does not meet the required accuracy level for this task. VGG16-B1 is excluded due to its lower performance.

Based on the combined evaluation of accuracy, stability, and deployment efficiency, **EfficientNetB0-S6** is selected as the final model for robot deployment.

### Implementing the Capture Node

The capture node was implemented to subscribe to the robot’s camera topic `/depth_cam/rgb/image_raw` and save incoming images to a specified directory. The node uses `cv_bridge` to convert ROS image messages to OpenCV images and writes each frame as a JPEG file.

The implementation was written in `capture_sub.py` and placed inside the `capture_topic` package. The key components of the node are as follows:

- Subscribes to `/depth_cam/rgb/image_raw`
- Converts incoming ROS Image messages to OpenCV format
- Saves images to the directory specified by the `--save_path` argument
- Limits the number of saved images based on `--num_max` to manage storage
- Displays the live image stream using OpenCV windowing

The main components of the implementation:

```python
import sys
import os
import rclpy
from rclpy.node import Node
from sensor_msgs.msg import Image
import cv2
from cv_bridge import CvBridge
import torch
import numpy as np
import PIL.Image as PIL_Image
from rclpy.parameter import Parameter
import argparse


class CamCapture(Node):
    def __init__(self,name, save_path, num_max):
        super().__init__(name)
        self.cam_subscription=self.create_subscription(Image, '/depth_cam/rgb/image_raw', self.image_callback, 1)
        self.cv_bridge=CvBridge()
        self.save_path=save_path
        os.makedirs(self.save_path, exist_ok=True)
        self.num_max = num_max
        self.save_id = 0
        
    def image_callback(self,msg):
        image_bgr=self.cv_bridge.imgmsg_to_cv2(msg,"bgr8")
        image_rgb = cv2.cvtColor(image_bgr, cv2.COLOR_BGR2RGB)
        cv2.imshow("capture rgb image", image_rgb)
        cv2.imwrite(os.path.join(self.save_path, f'image{self.save_id}.jpg'), image_bgr)
        self.save_id = (self.save_id + 1) % self.num_max
        cv2.waitKey(1)


def main(args=None):
    rclpy.init(args=args)
    parser = argparse.ArgumentParser(description='Save images from ROS2 topic')
    parser.add_argument('--save_path', type=str, default='~/script test/images')
    parser.add_argument('--num_max', type=int, default=1)
    parsed_args, unknown = parser.parse_known_args()
    node = CamCapture("capture_sub", parsed_args.save_path, parsed_args.num_max)
    rclpy.spin(node)
    rclpy.shutdown()

## Task 2: Fine-Tuning and Testing the Selected Classifier

For this task, we selected the EfficientNetB0-S6 model, which was identified as the best-performing classifier during Phase 2. The objective of this phase is to adapt and evaluate the model on images captured using the robot's onboard camera.

The process is divided into three stages:

1. The classifier is first fine-tuned on the complete Phase 2 dataset (all 23 classes). This step ensures that the model has been optimized on the full available data before evaluating its performance on the new robot-captured images.

2. The model is then tested on the test set of images captured from the robot camera (three selected classes). The purpose of this evaluation is to observe any potential domain shift or degradation in performance when moving from Phase 2 images to robot-captured images.

3. Finally, the classifier is further fine-tuned using the fine-tuning set of robot images (three selected classes), and tested again on the robot test set. This step allows the model to better adapt to the domain characteristics of robot-captured images, with the expectation of improved classification performance on these images.

In this section, we document the training and evaluation process at each stage, compare results to Phase 2 performance, and analyze the impact of fine-tuning on robot images.

The following class defines a custom PyTorch dataset to load images captured by the robot camera. It reads images organized in class-specific subfolders, applies the specified preprocessing transforms, and prepares the data for fine-tuning and testing of the classifier.

`````python
from torch.utils.data import Dataset
from PIL import Image
import os

class RobotCatDataset(Dataset):
    def __init__(self, root_dir, transform=None):
        self.root_dir = root_dir
        self.transform = transform
        self.classes = sorted(os.listdir(root_dir))
        self.image_paths = []
        self.labels = []

        for idx, class_name in enumerate(self.classes):
            class_folder = os.path.join(root_dir, class_name)
            for img_name in os.listdir(class_folder):
                if img_name.lower().endswith(('.jpg', '.jpeg', '.png')):
                    self.image_paths.append(os.path.join(class_folder, img_name))
                    self.labels.append(idx)

    def __len__(self):
        return len(self.image_paths)

    def __getitem__(self, idx):
        img_path = self.image_paths[idx]
        image = Image.open(img_path).convert('RGB')
        label = self.labels[idx]

        if self.transform:
            image = self.transform(image)

        return image, label
`````

`````python
# Robot Dataset Path
robot_data_path = 'Images and labels/final images'

# Load full robot dataset using existing data_transforms
robot_dataset = RobotCatDataset(robot_data_path, transform=data_transforms['train'])

# Split (example: 60/40)
train_size = int(0.6 * len(robot_dataset))
test_size = len(robot_dataset) - train_size

robot_train_data, robot_test_data = random_split(robot_dataset, [train_size, test_size])

# Dataloaders (also use same batch size as before if you want — e.g., batch_size = 8 or 16)
robot_train_loader = DataLoader(robot_train_data, batch_size=8, shuffle=True)
robot_test_loader = DataLoader(robot_test_data, batch_size=8, shuffle=False)

# Print summary
print("Robot Dataset Classes:", robot_dataset.classes)
print(f"Robot Fine-Tune Set: {len(robot_train_data)}")
print(f"Robot Test Set: {len(robot_test_data)}")
`````

The following function evaluates the trained classifier on the robot test set. It computes overall accuracy, generates a confusion matrix for the three selected classes, and provides a detailed classification report including precision, recall, and F1-scores.

`````python
def test_model(model, test_loader, class_names):
    model.eval()
    model.to(device)
    correct = 0
    total = 0
    all_preds = []
    all_labels = []

    with torch.no_grad():
        for inputs, labels in test_loader:
            inputs, labels = inputs.to(device), labels.to(device)
            outputs = model(inputs)
            _, preds = torch.max(outputs, 1)
            total += labels.size(0)
            correct += (preds == labels).sum().item()

            all_preds.extend(preds.cpu().numpy())
            all_labels.extend(labels.cpu().numpy())

    acc = 100 * correct / total
    print(f"\n[Test on Robot Test Set] Accuracy = {acc:.2f}%")

    # Confusion matrix
    cm = confusion_matrix(all_labels, all_preds)
    plt.figure(figsize=(6,5))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                xticklabels=class_names, yticklabels=class_names)
    plt.xlabel("Predicted")
    plt.ylabel("True")
    plt.title("Confusion Matrix - Robot Test Set")
    plt.show()

    print("\nClassification Report:")
    print(classification_report(all_labels, all_preds, target_names=class_names, labels=[0,1,2]))

    return acc
`````

This degradation in performance can be explained by the domain gap between the original Phase 2 training data and the robot camera images. The Phase 2 dataset consists of high-quality images from online sources, while the robot images are captured in varying lighting conditions, with different backgrounds, resolutions, and potential noise introduced by the robot’s camera system.

The significant drop in accuracy compared to Phase 2 results confirms that the model, although effective on the original dataset, does not generalize well to the new domain. These findings clearly justify the need for additional fine-tuning using robot-captured images to adapt the model to the characteristics of this domain and improve classification performance on the robot.

`````python
# Robot Dataset Path
robot_data_path = 'Images and labels/final images'

# Load full robot dataset using existing data_transforms
robot_dataset = RobotCatDataset(robot_data_path, transform=data_transforms['train'])

# Split (example: 60/40)
train_size = int(0.6 * len(robot_dataset))
test_size = len(robot_dataset) - train_size

robot_train_data, robot_test_data = random_split(robot_dataset, [train_size, test_size])

# Dataloaders
robot_train_loader = DataLoader(robot_train_data, batch_size=8, shuffle=True)
robot_test_loader = DataLoader(robot_test_data, batch_size=8, shuffle=False)

# Print summary
print("Robot Dataset Classes:", robot_dataset.classes)
print(f"Robot Fine-Tune Set: {len(robot_train_data)}")
print(f"Robot Test Set: {len(robot_test_data)}")
`````

The following training loop performs final fine-tuning of the EfficientNetB0-S6 model using the robot-captured fine-tuning set. The optimizer and learning rate scheduler are configured to adapt the model to the robot image domain. Validation accuracy on the robot test set is tracked after each epoch to monitor performance improvements during this adaptation process.

`````python
# Prepare optimizer & scheduler for final robot fine-tune
optimizer_robot = optim.Adam(filter(lambda p: p.requires_grad, effnet_model.parameters()), lr=0.001)
scheduler_robot = optim.lr_scheduler.ReduceLROnPlateau(optimizer_robot, patience=3, factor=0.3)

# Fine-tune loop
def fine_tune_model(model, train_loader, val_loader, criterion, optimizer, scheduler=None, epochs=5):
    model.to(device)
    val_acc = []
    best_acc = 0

    for epoch in range(epochs):
        model.train()
        for inputs, labels in train_loader:
            inputs, labels = inputs.to(device), labels.to(device)
            optimizer.zero_grad()
            outputs = model(inputs)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()

        # Validation on robot test set
        model.eval()
        correct, total = 0, 0
        with torch.no_grad():
            for inputs, labels in val_loader:
                inputs, labels = inputs.to(device), labels.to(device)
                outputs = model(inputs)
                _, preds = torch.max(outputs, 1)
                total += labels.size(0)
                correct += (preds == labels).sum().item()

        acc = 100 * correct / total
        val_acc.append(acc)

        if acc > best_acc:
            best_acc = acc

        if scheduler:
            if isinstance(scheduler, torch.optim.lr_scheduler.ReduceLROnPlateau):
                scheduler.step(acc)
            else:
                scheduler.step()

        print(f"[Final Robot Fine-Tune] Epoch {epoch+1}: Val Accuracy = {acc:.2f}%")

    print(f"[Final Robot Fine-Tune] Best Val Accuracy = {best_acc:.2f}%")
    return val_acc
`````

The following script implements a live webcam classifier using the final fine-tuned EfficientNetB0-S6 model. It captures frames from the robot’s camera stream, performs real-time classification, and overlays the predicted class label and confidence on the video feed. This provides a direct demonstration of the model’s performance during live deployment.


```python
# live_classifier.py

import cv2
import torch
import torch.nn.functional as F
from torchvision import models, transforms
import argparse
import time
from PIL import Image

# ✅ Load EfficientNetB0 model
def load_efficientnet_model(model_path, num_classes):
    model = models.efficientnet_b0(pretrained=False)
    model.classifier[1] = torch.nn.Linear(model.classifier[1].in_features, num_classes)
    model.load_state_dict(torch.load(model_path, map_location='cpu'))
    model.eval()
    return model

# ✅ Main live classifier
def run_live_classifier(model, class_names, device, camera_id=0):
    # Open webcam
    cap = cv2.VideoCapture(camera_id)
    if not cap.isOpened():
        print(f"Error: Could not open webcam (ID {camera_id})")
        return
    
    print(f"Webcam opened successfully (ID {camera_id})")
    print("Press 'q' to quit")

    # Define transforms (match your validation transforms)
    transform = transforms.Compose([
        transforms.Resize((224, 224)),
        transforms.ToTensor(),
        transforms.Normalize(mean=[0.485, 0.456, 0.406],
                             std=[0.229, 0.224, 0.225])
    ])

    try:
        while True:
            ret, frame = cap.read()
            if not ret:
                print("Failed to grab frame")
                break

            # Convert to PIL Image
            frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
            pil_image = Image.fromarray(frame_rgb)

            # Preprocess
            input_tensor = transform(pil_image).unsqueeze(0).to(device)

            # Inference
            with torch.no_grad():
                start_time = time.time()
                output = model(input_tensor)
                inference_time = time.time() - start_time

                probs = F.softmax(output, dim=1)
                confidence, pred_idx = torch.max(probs, 1)

            class_name = class_names[pred_idx.item()]
            confidence_pct = confidence.item() * 100

            # Display result on frame
            display_text = f"{class_name} ({confidence_pct:.2f}%) - {inference_time*1000:.1f}ms"
            cv2.putText(frame, display_text, (10, 30),
                        cv2.FONT_HERSHEY_SIMPLEX, 0.8, (0, 255, 0), 2)

            # Show frame
            cv2.imshow('Live Cat Breed Classifier', frame)

            # Press 'q' to quit
            if cv2.waitKey(1) & 0xFF == ord('q'):
                break

    except KeyboardInterrupt:
        print("Interrupted by user")

    finally:
        cap.release()
        cv2.destroyAllWindows()
        print("Webcam closed")

# ✅ Main entry point
if __name__ == '__main__':
    parser = argparse.ArgumentParser(description='Live Cat Breed Classifier')
    parser.add_argument('--model', type=str, required=True,
                        help='Path to model .pt file (e.g., efficientnet_s4_best.pt)')
    parser.add_argument('--class_file', type=str, required=True,
                        help='Path to classes.txt file')
    parser.add_argument('--camera', type=int, default=0,
                        help='Webcam camera ID (default: 0)')
    args = parser.parse_args()

    # Load class names
    with open(args.class_file, 'r') as f:
        class_names = [line.strip() for line in f.readlines()]
    
    num_classes = len(class_names)
    print(f"Loaded {num_classes} classes")

    # Device
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    print(f"Using device: {device}")

    # Load model
    model = load_efficientnet_model(args.model, num_classes)
    model.to(device)

    # Run live classifier
    run_live_classifier(model, class_names, device, camera_id=args.camera)

### Implementing the Classifier Trigger Script

To complete the robot deployment, a final script (`classifier_trigger.py`) was implemented to enable real-time monitoring and control of the robot based on live classification results.

This script performs the following functions:

1. **Model Loading**  
   The fine-tuned EfficientNetB0-S6 model is loaded from `efficientnet_s4_best.pt`. The model is configured for 23 classes, corresponding to the original Phase 2 class labels.

2. **Preprocessing Setup**  
   The script defines the required image preprocessing pipeline (resize to 224x224, normalization) to ensure that incoming images are processed consistently with the model’s training.

3. **Robot Control Initialization**  
   Before starting the monitoring loop, the robot is commanded to begin moving using the `ros2 run machine_move move` command with a defined linear velocity of 0.01. This establishes the robot’s baseline behavior of moving forward.

4. **Image Monitoring Loop**  
   The script continuously monitors a designated folder (`~/script test/images`), where images captured from the robot camera are saved. Each new image is processed and classified by the model.

5. **Classification and Action Trigger**  
   For each image, the model outputs a predicted class and corresponding confidence score (computed using softmax). If the confidence for the predicted class exceeds a defined threshold (45%), and if the predicted class is one of the cat breeds, the script triggers a stop command:

   ```bash
   ros2 run machine_move stop
