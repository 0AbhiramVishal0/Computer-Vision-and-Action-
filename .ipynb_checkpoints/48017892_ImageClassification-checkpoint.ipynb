# Introduction

This project focuses on fine-grained image classification of 23 different cat breeds. Fine-grained classification is challenging because all breeds belong to the same category (Cats) and have very small visual differences.

### Dataset Overview:
The dataset was collected together by our group of three members (Sadia, Subhana, Abhi). Images were collected from:
- CFA official website
- Google search
- Instagram
- Petfinder.com
- Cattime.com
- Personal pet photos

After collection, the dataset was divided equally among all 3 members. Each member got their own images for all 23 breeds. The dataset is organized inside a folder called `Data` with separate folders for Sadia, Subhana, and Abhi.

---

### Cat Breeds Used:
Abyssinian, American Shorthair, Balinese, Bengal, Birman, British Shorthair, Burmese, Cornish Rex, Devon Rex, Egyptian Mau, Exotic, Japanese Bobtail, Maine Coon Cat, Norwegian Forest Cat, Oriental, Persian, Ragdoll, Russian Blue, Scottish Fold, Selkirk Rex, Siamese, Siberian, Sphynx.

---

### CNN Models Used:
In this notebook, I am using the my collected dataset. For classification, I am testing and comparing two pretrained CNN models:

- **VGG16** – Simple and popular model  
- **ResNet50** – Deep model with residual connections  

Both models are pretrained on ImageNet and fine-tuned on this dataset.


```python
# Basic Libraries
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

# PyTorch Libraries
import torch
import torchvision
from torchvision import datasets, transforms, models
from torch.utils.data import DataLoader, random_split
import torch.nn as nn
import torch.optim as optim

# Sklearn for Evaluation
from sklearn.metrics import confusion_matrix, classification_report
from PIL import Image
import random
```

## 1. Data set Collection
Loading Abhi's dataset and exploring the data.

# 2. Dataset Split
- Dataset is divided into Train, Validation, and Test sets.
- Train set is used to train the model.
- Validation set is used to tune the model during training.
- Test set is used to check the final performance of the model.
- The split ratio used is 60% Train, 10% Validation, and rest for test 

```python
from torch.utils.data import DataLoader, random_split

# Load Dataset with Transformations
dataset = datasets.ImageFolder(root=base_path, transform=data_transforms['train'])

# Split Sizes
train_size = int(0.6 * len(dataset))
val_size = int(0.1 * len(dataset))
test_size = len(dataset) - train_size - val_size

# Split the Dataset
train_data, val_data, test_data = random_split(dataset, [train_size, val_size, test_size])

print(f"Train Images: {len(train_data)}")
print(f"Validation Images: {len(val_data)}")
print(f"Test Images: {len(test_data)}")
```

## Data Transformations
- Image transformations are applied to improve model learning.
- Resize ensures all images are of the same size.
- Data Augmentation like horizontal flip & rotation adds more variety.
- Normalization helps models trained on ImageNet to work better with our dataset.

```python
from torchvision import transforms

# Define Transformations
data_transforms = {
    'train': transforms.Compose([
        transforms.Resize((224, 224)),
        transforms.RandomHorizontalFlip(),
        transforms.RandomRotation(10),
        transforms.ToTensor(),
        transforms.Normalize([0.485, 0.456, 0.406],
                             [0.229, 0.224, 0.225])
    ]),
    'val': transforms.Compose([
        transforms.Resize((224, 224)),
        transforms.ToTensor(),
        transforms.Normalize([0.485, 0.456, 0.406],
                             [0.229, 0.224, 0.225])
    ]),
}
```

## Data Loaders
- DataLoaders help in loading data in batches.
- They improve training efficiency.
- Train loader shuffles data for better learning.
- Validation and Test loaders keep data order fixed.


```python
# Create DataLoaders
batch_size = 32

train_loader = DataLoader(train_data, batch_size=batch_size, shuffle=True)
val_loader = DataLoader(val_data, batch_size=batch_size, shuffle=False)
test_loader = DataLoader(test_data, batch_size=batch_size, shuffle=False)

print("DataLoaders created successfully!")
```

# 3. Pretrained Model Selection
- We have selected two popular CNN models: VGG16 and ResNet50.
- VGG16 is a simple and widely used deep learning model for image classification tasks.
- ResNet50 is a deeper model with skip connections, making it more effective for fine-grained classification problems.
- Both models are pretrained on ImageNet and suitable for transfer learning on our dataset.

```python
from torchvision import models

# Load Pretrained VGG16
vgg_model = models.vgg16(pretrained=True)

# Load Pretrained ResNet50
resnet_model = models.resnet50(pretrained=True)

print("VGG16 and ResNet50 models loaded successfully!")
```

## Model Modification
- The original models were trained to classify 1000 classes.
- We modified the final layer to classify 23 cat breeds in our dataset.
- Feature extraction layers are frozen to use pretrained knowledge.
- Only the final classification layer will be trained.

```python
import torch.nn as nn

# Modify VGG16 Final Layer
for param in vgg_model.features.parameters():
    param.requires_grad = False

vgg_model.classifier[6] = nn.Linear(4096, 23)

# Modify ResNet50 Final Layer
for param in resnet_model.parameters():
    param.requires_grad = False

resnet_model.fc = nn.Linear(resnet_model.fc.in_features, 23)

print("Final layers of both models modified for 23 classes.")
```

## Loss Function & Optimizer
- CrossEntropyLoss is used as it works well for multi-class classification.
- Adam optimizer is used to update only the final layer parameters of both models.
- Learning rate is set to 0.001 for stable learning.

```python
# Loss Function
criterion = nn.CrossEntropyLoss()

# Optimizer for VGG16
optimizer_vgg = torch.optim.Adam(vgg_model.classifier.parameters(), lr=0.001)

# Optimizer for ResNet50
optimizer_resnet = torch.optim.Adam(resnet_model.fc.parameters(), lr=0.001)

print("Loss function and optimizers set successfully for both models.")
```

## Training Function
- This function trains the model for a given number of epochs.
- It calculates training loss after every epoch.
- Model is set to training mode before each epoch.

```python
def train_model(model, train_loader, val_loader, criterion, optimizer, epochs=10):
    for epoch in range(epochs):
        model.train()
        running_loss = 0.0
        
        for inputs, labels in train_loader:
            inputs, labels = inputs.to('cuda' if torch.cuda.is_available() else 'cpu'), labels.to('cuda' if torch.cuda.is_available() else 'cpu')

            optimizer.zero_grad()
            outputs = model(inputs)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()

            running_loss += loss.item()

        print(f"Epoch [{epoch+1}/{epochs}] - Training Loss: {running_loss/len(train_loader):.4f}")

    print("Training Completed!")
```

## Training VGG16
- We are using a pretrained VGG16 model with weights trained on ImageNet dataset.
- Pretrained weights help the model to extract general image features like edges, patterns, and textures.
- Only the final layer is trained on our dataset, which saves time and prevents overfitting on small data.
- This method is called Transfer Learning.

```python
# Move model to device
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
vgg_model = vgg_model.to(device)

# Train VGG16
train_model(vgg_model, train_loader, val_loader, criterion, optimizer_vgg, epochs=10)
```

## Evaluation of VGG16
- The evaluation is done on the test dataset.
- Accuracy, classification report, and confusion matrix are used to measure performance.

```python
def evaluate_model(model, test_loader):
    model.eval()
    correct = 0
    total = 0
    all_preds = []
    all_labels = []

    with torch.no_grad():
        for inputs, labels in test_loader:
            inputs, labels = inputs.to(device), labels.to(device)
            outputs = model(inputs)
            _, preds = torch.max(outputs, 1)

            all_preds.extend(preds.cpu().numpy())
            all_labels.extend(labels.cpu().numpy())

            correct += (preds == labels).sum().item()
            total += labels.size(0)

    accuracy = 100 * correct / total
    print(f"Test Accuracy: {accuracy:.2f}%")
    return all_labels, all_preds
```

## VGG16 Evaluation Results
Evaluating the performance of VGG16 model on test data.

```python
# Evaluate VGG16
true_labels_vgg, predictions_vgg = evaluate_model(vgg_model, test_loader)
```

## VGG16 Performance Metrics
Displaying classification report and confusion matrix for VGG16 model.

```python
# Classification Report
print(classification_report(true_labels_vgg, predictions_vgg, target_names=dataset.classes))

# Confusion Matrix
cm_vgg = confusion_matrix(true_labels_vgg, predictions_vgg)

plt.figure(figsize=(15, 10))
sns.heatmap(cm_vgg, annot=True, fmt='d', cmap='Blues',
            xticklabels=dataset.classes,
            yticklabels=dataset.classes)

plt.title("Confusion Matrix - VGG16")
plt.xlabel("Predicted")
plt.ylabel("True")
plt.show()
```

## Training ResNet50
Training ResNet50 model using the same dataset and settings.

```python
# Move ResNet50 model to device
resnet_model = resnet_model.to(device)

# Train ResNet50
train_model(resnet_model, train_loader, val_loader, criterion, optimizer_resnet, epochs=10)
```

## Evaluation of ResNet50
Evaluating the performance of ResNet50 model on test data.

```python
# Evaluate ResNet50
true_labels_resnet, predictions_resnet = evaluate_model(resnet_model, test_loader)
```

## ResNet50 Performance Metrics
Displaying classification report and confusion matrix for ResNet50 model.

```python
# Classification Report
print(classification_report(true_labels_resnet, predictions_resnet, target_names=dataset.classes))

# Confusion Matrix
cm_resnet = confusion_matrix(true_labels_resnet, predictions_resnet)

plt.figure(figsize=(15, 10))
sns.heatmap(cm_resnet, annot=True, fmt='d', cmap='Greens',
            xticklabels=dataset.classes,
            yticklabels=dataset.classes)

plt.title("Confusion Matrix - ResNet50")
plt.xlabel("Predicted")
plt.ylabel("True")
plt.show()
```

## Final Comparison
Comparing the performance of VGG16 and ResNet50 based on test accuracy.

```python
# Accuracy Calculation
def calculate_accuracy(true, preds):
    correct = sum(np.array(true) == np.array(preds))
    total = len(true)
    return 100 * correct / total

vgg_acc = calculate_accuracy(true_labels_vgg, predictions_vgg)
resnet_acc = calculate_accuracy(true_labels_resnet, predictions_resnet)

print(f"VGG16 Test Accuracy   : {vgg_acc:.2f}%")
print(f"ResNet50 Test Accuracy: {resnet_acc:.2f}%")
```

```python
# Bar Plot of Accuracy Comparison
plt.figure(figsize=(6, 6))
models = ['VGG16', 'ResNet50']
accuracies = [vgg_acc, resnet_acc]

sns.barplot(x=models, y=accuracies, palette='pastel')
plt.ylim(0, 100)
plt.ylabel('Accuracy (%)')
plt.title('VGG16 vs ResNet50 Accuracy Comparison')
plt.show()
```

## Model Evaluation & Comparison Analysis

Based on our experiments, the performance comparison between VGG16 and ResNet50 models is clearly visible.

### Test Accuracy:
- VGG16 Test Accuracy: 55.6%
- ResNet50 Test Accuracy: 71.0%

---

### Key Observations:

- ResNet50 outperformed VGG16 in every evaluation metric.
- VGG16 struggled with fine-grained details, especially for breeds having similar features like Burmese, British Shorthair, and Devon Rex.
- ResNet50 gave a higher macro average F1-Score of 0.67 compared to VGG16's 0.49.
- The weighted average F1-Score of ResNet50 is 0.71, which is higher than VGG16's 0.51.

---

### Strengths of VGG16:
- Performed decently on clear, front-view images.
- Performed well for breeds like American Shorthair (Precision: 1.00), Devon Rex (Precision: 1.00), and Persian (Precision: 0.78).

### Weaknesses of VGG16:
- Struggled heavily with rare views, side views, or poor-quality images.
- Many breeds like Balinese, Bengal, Oriental had very low recall (0.00).

---

### Strengths of ResNet50:
- Handled variation in images (pose, view, lighting) much better.
- Achieved strong recall for most breeds.
- Performed well for difficult breeds like Burmese (F1-score: 0.89), Siamese (0.93), Ragdoll (0.91), and Russian Blue (0.95).

### Weaknesses of ResNet50:
- Still struggled for a few classes like Exotic (F1-score: 0.22) and Bengal (0.40), mainly due to lack of enough diverse images.

---

### Final Conclusion:
ResNet50 is clearly more suitable for fine-grained cat breed classification tasks compared to VGG16. The deeper architecture with skip connections helps ResNet50 extract more complex features, leading to better performance in handling subtle visual differences between cat breeds.
